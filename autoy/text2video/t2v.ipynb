## Setting functions

```python
import pandas as pd
import cv2
import random
import datetime
import numpy as np
from PIL import ImageFont, ImageDraw, Image

# t2v class
class T2V:
    def __init__(self,
                 arg_resolution, bg_randomness=True):
        self.bg_randomness = bg_randomness
        self.height = resolution_dict[arg_resolution][0]
        self.width = resolution_dict[arg_resolution][1]
        print('resolution ì„¸íŒ…:', arg_resolution)
        print('height:', self.height)
        print('width:', self.width)
        
        self.frame_array = []
    
    # settings of text
    def setting_text(self, arg_font_defalut, arg_font_size_default, 
                     arg_font_word, arg_font_size_word):
        font = font_dict[arg_font_defalut]
        font_word = font_dict[arg_font_word]
        font_size_defalut = arg_font_size_default
        font_size_word = arg_font_size_word
        self.font_pil = ImageFont.truetype(font=font,
                                           size=font_size_defalut)
        self.font_pil_word = ImageFont.truetype(font=font_word,
                                                size=font_size_word)
        print("ê¸°ë³¸ Font ì„¸íŒ…:", font, font_size_word)
        print("ë‹¨ì–´ Font ì„¸íŒ…:", font_word, font_size_word)
        print('\n[START] ì˜ìƒ ë§Œë“¤ê¸°\n')

    # setting of video
    def making_imageAraay(self, arg_word_dict):

        if self.bg_randomness == True:
            randNum = random.sample(range(0,6),1)
            self.bg_img_path = background_dict[randNum[0]]
            print("Video ë°°ê²½ random ì„¸íŒ…:", self.bg_randomness, randNum[0])
        else:
            self.bg_img_path = background_dict[999]

        xy_title = (self.width*(0.2/10), self.height*(1/20))
        xy_logo = (self.width*(8.5/10), self.height*(1/20))
        xy_progress = (self.width*(8.8/10), self.height*(1.3/10))
        xy_world = (self.width*(1/2), self.height*(2.2/5))
        xy_meaning = (self.width*(1/2), self.height*(2.6/5))
        xy_subject = (self.width*(1/2), self.height*(2.5/5))
        xy_sentence = (self.width*(1/2), self.height*(4/5))

        words_dict_target = arg_word_dict 
        df = pd.DataFrame(words_dict_target, columns=columns)

        for i in range(len(words_dict_target)):
            word = df[i:i+1]['ë‹¨ì–´'].values[0]
            meaning = df[i:i+1]['ëœ»'].values[0]
            sentence = df[i:i+1]['ë¬¸ì¥ ì˜ˆì‹œ'].values[0]
            progress = str(i+1) + "/" + str(len(words_dict_target))

            if len(self.bg_img_path) == 0:
                bg_img = np.full(shape=(self.height,self.width,3), 
                                    fill_value=255, dtype=np.uint8)
            else:
                bg_img = cv2.imread(filename=self.bg_img_path)

            bg_img_pil = Image.fromarray(bg_img)
            draw=ImageDraw.Draw(bg_img_pil)

            draw.text(xy=xy_world, text=word,
                fill=(0,0,0), font=self.font_pil_word,
                anchor='mm', align='center')
            draw.text(xy=xy_meaning, text=meaning,
                fill=(0,0,0), font=self.font_pil_word,
                anchor='mm', align='center')
            # draw.text(xy=xy_sentence, text=sentence,
            #     fill=(0,0,0), font=font_pil,
            #     anchor='mm', align='center')
            draw.text(xy=xy_progress, text=progress,
                fill=(0,0,0), font=self.font_pil,
                anchor='mm', align='center')

            bg_img_array = np.array(bg_img_pil)
            self.frame_array.append(bg_img_array)
            if (i+1)%5 == 0:
                print("ì™„ë£Œ frame ë²ˆí˜¸:", i+1, "/", len(words_dict_target))
            elif i == len(words_dict_target)-1:
                print("ì™„ë£Œ frame ë²ˆí˜¸:", i+1, "/", len(words_dict_target))

            word = None
            meaning = None
            sentence = None
            progress = None

        print("ì´ frame ìˆ˜:", len(self.frame_array))
        return self.frame_array
    
    # save video file
    def saveVideo(self, arg_subject, start_num, arg_fps, arg_fourcc):
        now = datetime.datetime.now()
        now = now.strftime('%Y%m%d%H%M%S')
        filename = './output/' + arg_subject + '_' + start_num + '_' + now + arg_fourcc

        out_video = cv2.VideoWriter(filename=filename, 
                                    fourcc=cv2.VideoWriter_fourcc(*fourcc_dict[arg_fourcc]), 
                                    fps=arg_fps, frameSize=(self.width, self.height))

        for i in range(len(self.frame_array)):
            out_video.write(self.frame_array[i])
            out_video.write(self.frame_array[i])
            out_video.write(self.frame_array[i])

        out_video.release()
        self.frame_array = []
        print("[COMPLETE] Video ìƒì„± ì™„ë£Œ:", filename + '\n')
```
