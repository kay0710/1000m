### Options
- target_word_dict: 사용할 단어 모음 선택
    - 추가 형식 >> '사용자 정의 이름': 단어 모음 이름(변수명)
- resolution_dict: 생성할 영상 사이즈 옵션들을 저장
    - 추가 형식 >> '사이즈 이름': (세로, 가로, 3)
- location_dict: 영상 사이즈 기준으로 content의 위치 옵션들 저장
    - 사용법: location_dict['yshorts_1500']['xy_title'][0]
- font_dict: 사용할 font 옵션들을 저장
    - 추가 형식 >> 'font 이름': 'font 파일 경로'
- background_dict: 영상의 배경으로 사용할 옵션들을 저장
    - 추가 형식 >> 번호: '배경 파일 경로'
- fourcc_dict: 비디오 생성 코덱 설정
    - 추가 형식 >> '.확장자': 코덱명(opencv)

```python
target_word_dict = {'travel50': travel_word50,
                    'high_school50': hs_word50,

}

resolution_dict = {'square_500': (500,500,3),
                   'square_1000': (1000,1000,3),
                   'yshorts_1500': (1500,843,3),
                   'yshorts_1900': (1920,1080,3),
}

location_dict = {'yshorts_1500': {'xy_title': ((0.2/10), (1/20)),
                                'xy_logo': ((8.5/10), (1/20)),
                                'xy_progress': ((8.8/10), (1.3/10)),
                                'xy_world': ((1/2), (2.2/5)),
                                'xy_meaning': ((1/2), (2.6/5)),
                                'xy_subject': ((1/2), (2.5/5)),
                                'xy_sentence': ((1/2), (4/5))}
}

font_dict = {'marunuri_light': './fonts/마루 부리/MaruBuriTTF/MaruBuri-Light.ttf',
             'nanumGothic': './fonts/나눔 글꼴/나눔고딕/NanumFontSetup_TTF_GOTHIC/NanumGothic.ttf',
             'maruburi_bold': './fonts/마루 부리/MaruBuriTTF/MaruBuri-Bold.ttf',
    
}

background_dict = {0: './background/can Talk 1.png',
                   1: './background/can Talk 2.png',
                   2: './background/can Talk 3.png',
                   3: './background/can Talk 4.png',
                   4: './background/can Talk 5.png',
                   5: './background/can Talk 6.png',
                   999: '' # fixed background

}

fourcc_dict = {'.avi': 'DIVX',
            #    '.mp4': 'MP4S',
               '.mp4': 'MP4V',
               '.wmv': 'WMV1'

}
```

## Setting functions

```python
import pandas as pd
import cv2
import random
import datetime
import numpy as np
from PIL import ImageFont, ImageDraw, Image

# t2v class
class T2V:
    def __init__(self,
                 arg_resolution, bg_randomness=True):
        self.bg_randomness = bg_randomness
        self.height = resolution_dict[arg_resolution][0]
        self.width = resolution_dict[arg_resolution][1]
        print('resolution 세팅:', arg_resolution)
        print('height:', self.height)
        print('width:', self.width)
        
        self.frame_array = []
    
    # settings of text
    def setting_text(self, arg_font_defalut, arg_font_size_default, 
                     arg_font_word, arg_font_size_word):
        font = font_dict[arg_font_defalut]
        font_word = font_dict[arg_font_word]
        font_size_defalut = arg_font_size_default
        font_size_word = arg_font_size_word
        self.font_pil = ImageFont.truetype(font=font,
                                           size=font_size_defalut)
        self.font_pil_word = ImageFont.truetype(font=font_word,
                                                size=font_size_word)
        print("기본 Font 세팅:", font, font_size_word)
        print("단어 Font 세팅:", font_word, font_size_word)
        print('\n[START] 영상 만들기\n')

    # setting of video
    def making_imageAraay(self, arg_word_dict):

        if self.bg_randomness == True:
            randNum = random.sample(range(0,6),1)
            self.bg_img_path = background_dict[randNum[0]]
            print("Video 배경 random 세팅:", self.bg_randomness, randNum[0])
        else:
            self.bg_img_path = background_dict[999]

        xy_title = (self.width*(0.2/10), self.height*(1/20))
        xy_logo = (self.width*(8.5/10), self.height*(1/20))
        xy_progress = (self.width*(8.8/10), self.height*(1.3/10))
        xy_world = (self.width*(1/2), self.height*(2.2/5))
        xy_meaning = (self.width*(1/2), self.height*(2.6/5))
        xy_subject = (self.width*(1/2), self.height*(2.5/5))
        xy_sentence = (self.width*(1/2), self.height*(4/5))

        words_dict_target = arg_word_dict 
        df = pd.DataFrame(words_dict_target, columns=columns)

        for i in range(len(words_dict_target)):
            word = df[i:i+1]['단어'].values[0]
            meaning = df[i:i+1]['뜻'].values[0]
            sentence = df[i:i+1]['문장 예시'].values[0]
            progress = str(i+1) + "/" + str(len(words_dict_target))

            if len(self.bg_img_path) == 0:
                bg_img = np.full(shape=(self.height,self.width,3), 
                                    fill_value=255, dtype=np.uint8)
            else:
                bg_img = cv2.imread(filename=self.bg_img_path)

            bg_img_pil = Image.fromarray(bg_img)
            draw=ImageDraw.Draw(bg_img_pil)

            draw.text(xy=xy_world, text=word,
                fill=(0,0,0), font=self.font_pil_word,
                anchor='mm', align='center')
            draw.text(xy=xy_meaning, text=meaning,
                fill=(0,0,0), font=self.font_pil_word,
                anchor='mm', align='center')
            # draw.text(xy=xy_sentence, text=sentence,
            #     fill=(0,0,0), font=font_pil,
            #     anchor='mm', align='center')
            draw.text(xy=xy_progress, text=progress,
                fill=(0,0,0), font=self.font_pil,
                anchor='mm', align='center')

            bg_img_array = np.array(bg_img_pil)
            self.frame_array.append(bg_img_array)
            if (i+1)%5 == 0:
                print("완료 frame 번호:", i+1, "/", len(words_dict_target))
            elif i == len(words_dict_target)-1:
                print("완료 frame 번호:", i+1, "/", len(words_dict_target))

            word = None
            meaning = None
            sentence = None
            progress = None

        print("총 frame 수:", len(self.frame_array))
        return self.frame_array
    
    # save video file
    def saveVideo(self, arg_subject, start_num, arg_fps, arg_fourcc):
        now = datetime.datetime.now()
        now = now.strftime('%Y%m%d%H%M%S')
        filename = './output/' + arg_subject + '_' + start_num + '_' + now + arg_fourcc

        out_video = cv2.VideoWriter(filename=filename, 
                                    fourcc=cv2.VideoWriter_fourcc(*fourcc_dict[arg_fourcc]), 
                                    fps=arg_fps, frameSize=(self.width, self.height))

        for i in range(len(self.frame_array)):
            out_video.write(self.frame_array[i])
            out_video.write(self.frame_array[i])
            out_video.write(self.frame_array[i])

        out_video.release()
        self.frame_array = []
        print("[COMPLETE] Video 생성 완료:", filename + '\n')
```

## Making

```python
# settings
SUBJECT = 'Travel_EngWordsTop50' # using for video file name
dict_name = 'travel50' # name of using word dictionary
numOFword = 10 # number of words in video

RESOLUTION = 'yshorts_1500' # resolution of video
FONT_DEFAULT = 'nanumGothic' # font type of text (except words)
FONT_WORD = 'maruburi_bold' # font type of words
FONT_SIZE = 30 # font size of text (except words)
FONT_SIZE_WORD = 70 # font type of words

FPS = 0.6 # frames per seconds >> 0.6 = 5s/frame
FOURCC = '.mp4' # file type of video file

target_dict = target_word_dict[dict_name]

AUTOMATION = True # option / True or False
start_idx = 0 # start index of choosen word (## option for make 1 video)

print("영상 주제:", SUBJECT)
print("단어 범위:", numOFword)
print("사용할 항목:", columns)
print("생성 타입:", FOURCC)
```

### Do it!

```python
if AUTOMATION == True:
    t2v = T2V(RESOLUTION)
    t2v.setting_text(FONT_DEFAULT, FONT_SIZE, FONT_WORD, FONT_SIZE_WORD)

    for stx in range(0, ((len(target_dict)+numOFword)//numOFword)):
        etx = stx*10+numOFword
        TARGET_WORDS = target_dict[stx*10:etx]
        t2v.making_imageAraay(TARGET_WORDS)
        print("[Progress] 영상 생성 중:", stx+1, '/', (len(target_dict)//10)+1)
        t2v.saveVideo(arg_subject=SUBJECT, start_num=str(stx), arg_fps=FPS, arg_fourcc=FOURCC)
    tv2 = None
else:
    end_idx = start_idx+numOFword
    target_dict = target_word_dict[dict_name]
    TARGET_WORDS = target_dict[start_idx:end_idx]

    t2v = T2V(RESOLUTION)
    t2v.setting_text(FONT_DEFAULT, FONT_SIZE, FONT_WORD, FONT_SIZE_WORD)
    t2v.making_imageAraay(TARGET_WORDS)
    t2v.saveVideo(arg_subject=SUBJECT, start_num=str(start_idx), arg_fps=FPS, arg_fourcc=FOURCC)
    t2v = None
    
print('============================')
print("\n[END] 모든 영상 생성 완료\n\n")
```
